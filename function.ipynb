```python
## why function
num=24
if num%2==0:
    print("The number is even")
else:
    print("The number is odd")
```

```python
def even_or_odd(num):
    """"This function finds even or odd"""
    if num%2==0:
        print("The number is even")
    else:
        print("The number is odd")
```

```python
## Call this function
even_or_odd(24)
```

```python
## function with multiple parameters
def add(a,b):
    return a+b

result=add(2,4)
print(result)
```

```python
##  Default parameters
def greet(name):
    print(f"Hello {name} Welocome to the paradise")

greet("Abhinav")
```

```python
## Variable length arguments
## Positional and keywords arguments

def print_numbers(*args):
    for numbers in args:
        print(numbers)
```

```python
print_numbers(1,2,3,4,5,"args")
```

```python
## Keywords arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="Kanchan",age="21",country="India")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional arguments :{val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,5,"args",name="Kanchan",age="21",country="India")
```
