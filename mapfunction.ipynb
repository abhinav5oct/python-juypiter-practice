## Map() Function
The map() function applies a given function to all items in an input list (or any other iterable) and returns a map object (an iterator).This is particularly useful for transforming data in a list comprehensively.


```python
def square(x):
    return x*x
square(10)
```

```python
numbers=[1,2,3,4,5,6,7,8]

list(map(square,numbers))
```

```python
## Lambda function in map
numbers=[1,2,3,4,5,6,7,8]
list(map(lambda x:x*x,numbers))
```

```python
## Map multiple iterables
numbers1=[1,2,3]
numbers2=[4,5,6]

added_numbers=list(map(lambda x,y:x+y,numbers1,numbers2))
print(added_numbers)
```

```python
## map() to convert a list of string to intergers
str_numbers=['1','2','3','4','5']
int_numbers=list(map(int,str_numbers))
print(int_numbers)
```

```python
words=['apple','banana','cherry']
upper_word=list(map(str.upper,words))
print(upper_word)
```

```python
def get_name(person):
    return person['name']


people=[
    {'name':'kanchan','age':22},
    {'name':'jack','age':33}
]
list(map(get_name,people))
```

## Conclusion
The map() function is a powerful tool for applying transformation to iterable data structures.It can be used with reqular function,lambda functions,and even multiple iterables,providing a versatile approach to data processing inpython.By understanding and utilizing map(),you can write more efficient and readable code.
