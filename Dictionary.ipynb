## Introduction to Dictionaries

Dictionaries are unorderd collections of items. They store data in key-value pairs. Keys must be unique and  immutable (e.g, strings,numbers,or tuples),while values can be of any type.

```python
## creating Dictionaries

empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
empty_dict
```

```python
student={"name":"Abhinav","age":22,"grade":25}
print(student)
print(type(student))
```

```python
## Error in Dictionaries
##Single key is always used
student={"name":"Abhinav","age":22,"name":25}
print(student)
```

```python
### Accessing Values in Dictionaries
student={"name":"Abhinav","age":22,"grade":'A'}
print(student)
```

```python
# Accessing Dictionary elements

print(student['name'])
print(student['age'])
print(student['grade'])

## Accessing using get() method
print(student.get('name'))
print(student.get('age'))   
print(student.get('grade'))
print(student.get('last_name'),"Not Available")
```

```python
## Modifying Dictionaries
## Dictionaries are mutable,so you can add, udate or delete elemnents

print(student)
```

```python
student["age"]=33 ## update value for the key
print(student)
student["address"]="India"  ## added a new key and value
print(student)
```

```python
del student['grade'] ## delete key and value pair
print(student)
```

```python
## Dictionary Methods

keys=student.keys() ##get all the keys
print(keys)
values=student.values() ##get all values
print(values)

items=student.items() ## get all key value pairs
print(items)
```

```python
## Shallow copy of Dictionary
student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="Abhinav1"
print(student)
print(student_copy)
```

```python
student_copy1=student.copy() ## shallow copy 
print(student_copy1)
print(student)
```

```python
student["name"]="Abhinav2"
print(student_copy1)
print(student)
```

```python
## Iterating Over Dictionaries
## You can use loops to iterate over dictionaries,keys,values or items
for keys in student.keys():
    print(keys)

    ## iterating over values
    for values in student.values():
        print(values)
```

```python
## Iterate over key value pairs
for key,value in student.items():
    print(f"key:{key} and value:{value}")
```

```python
## Nested Dictionaries
students={
    "student1":{"name":"Abhinav","age":21},
    "student2":{"name":"Kanchan","age":21}
}
print(students)
```

```python
## Access nested dictionaries elements
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
students.items()
```

```python
## Iterating over nested dictionaries
for student_id,student_info in students.items():
     print(f"{student_id}:{student_info}")
     for key,value in student_info.items():
          print(f"{key}:{value}")  
```

```python
## Dictinorary comphrehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## Condationals dictionary comprehension
evens={x:x**2 for x in range(10) if x%2==0 }
print(evens)
```

```python
 ## practical examples
## use a dictionary to count the frequency of elemnets in list

numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```
